```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Reading CSV
Employment = pd.read_csv("../Dataset/Employment_to_Population.csv", encoding="ISO-8859-1")
Employment.head()
```

```python
Employment.set_index('Country', inplace=True)
Employment.head()
```

```python
#Counting Rows and Columns
Employment.columns
```

```python
# Dropping columns, in order to analyse from 2008 to 2018
Employment.drop(['Country Code','1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998',
       '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007'], axis=1, inplace=True)
Employment.head()
```

```python
# Filter Data for Canada, China, Philippines and India
Emp_compare= Employment.loc[["Canada", "China", "India", "Philippines"], :]
Emp_compare.head()
```

```python
#Transpose the Data to depict the Graph
Emp_df = Emp_compare.T
Emp_df.head(11)
```

```python
Emp_df.describe()
```

```python
# Plot the line
plt.figure(figsize=(10,5))
plt.title("Employment to Population from 2008 to 2018")
plt.plot(Emp_df)
plt.xlabel('Years')
plt.ylabel('Employment to Population Rate')
plt.legend(Emp_df, loc = 'best')
plt.show()
```
