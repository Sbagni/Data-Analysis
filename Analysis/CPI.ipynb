```python
import pandas as pd
import csv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
country_info = '../Dataset/cpi_data.csv'
value_indicators_df = pd.read_csv(country_info)
value_indicators_df['value'].max()
```

```python
# Filter the data so that only those sightings in the CPI are in a DataFrame
cpi_df = value_indicators_df.loc[value_indicators_df['factor'] == "cpi"][['countryidvalue','value','date']]
```

```python
cpi_df.head()
```

```python
x = cpi_df['date'].unique()
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)

plt.plot(x,cpi_df.loc[cpi_df['countryidvalue']=='Canada']['value'],linewidth=2,linestyle='--',label='Canada')
plt.plot(x,cpi_df.loc[cpi_df['countryidvalue']=='China']['value'],linewidth=2,label='China')
plt.plot(x,cpi_df.loc[cpi_df['countryidvalue']=='India']['value'],linewidth=2, color='r',label='India')
plt.plot(x,cpi_df.loc[cpi_df['countryidvalue']=='Philippines']['value'],linewidth=2,label='Philippines')
plt.title("Consumer Price Index of the 4 Countries\n")
plt.ylabel("CPI")
plt.xlabel("Year")
plt.grid(axis='y',linestyle='--')
plt.legend()
plt.show()
```
