```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv("./Resources/world_GDP.csv")
df.head()
```

```python
df.boxplot("GDP %", by="Country", figsize=(20, 10))
plt.savefig("Country_GDP%.png")
```

```python
group1 = df[df["Country"] == "Canada"]["GDP %"]
group2 = df[df["Country"] == "China"]["GDP %"]
group3 = df[df["Country"] == "India"]["GDP %"]
group4 = df[df["Country"] == "Philippines"]["GDP %"]
```

```python
stats.f_oneway(group1, group2, group3, group4)
```
